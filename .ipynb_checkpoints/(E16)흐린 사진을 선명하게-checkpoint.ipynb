# 16-5. Í∞ÄÏû• Ïâ¨Ïö¥ Super Resolution

Super ResolutionÏùÑ ÏàòÌñâÌïòÎäî Í∞ÄÏû• Ïâ¨Ïö¥ Î∞©ÏãùÏùÄ Interpolation  

Interpolation = ÏïåÎ†§ÏßÑ Í∞íÏùÑ Í∞ÄÏßÑ Îëê Ï†ê ÏÇ¨Ïù¥ Ïñ¥Îäê ÏßÄÏ†êÏùò Í∞íÏù¥ ÏñºÎßàÏùºÏßÄÎ•º Ï∂îÏ†ïÌïòÎäî Í∏∞Î≤ïÏùÑ ÏùòÎØ∏  

**Q5. Linear interpolationÍ≥º Bilinear interpolationÏùò Ï∞®Ïù¥Ï†êÏùÄ Î¨¥ÏóáÏù∏Í∞ÄÏöî?**  
>Linear interpolationÏùÑ 2Ï∞®ÏõêÏúºÎ°ú ÌôïÏû•Ìïú Í≤ÉÏù¥ bilinear interpolationÏûÖÎãàÎã§.  
Linear interpolationÏùò Í≤ΩÏö∞ 2Í∞úÏùò Í∞íÏùÑ Ïù¥Ïö©Ìï¥ ÏÉàÎ°úÏö¥ ÌîΩÏÖÄÏùÑ ÏòàÏ∏°ÌïòÎ©∞, bilinear interpolationÏùÄ 4Í∞úÏùò Í∞íÏùÑ Ïù¥Ïö©Ìï©ÎãàÎã§.

ÏòÅÏÉÅÏ≤òÎ¶¨ bilinear interpolation ÏòàÏ†ú : https://blog.naver.com/dic1224/220882679460

Ïù¥Î≤àÏóêÎäî opencv ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º Ïù¥Ïö©Ìï¥ interpolationÏùÑ Ï†ÅÏö©Ìï¥ Î¥ÖÏãúÎã§. Ï†ÅÏö©Ìï† Ïù¥ÎØ∏ÏßÄÎäî scikit-imageÏóêÏÑú Ï†úÍ≥µÌïòÎäî ÏòàÏ†ú Ïù¥ÎØ∏ÏßÄÎ•º ÏÇ¨Ïö©ÌïòÍ≤†ÏäµÎãàÎã§. ÎßåÏïΩ Îëê ÎùºÏù¥Î∏åÎü¨Î¶¨Í∞Ä ÏÑ§ÏπòÎêòÏñ¥ ÏûàÏßÄ ÏïäÎã§Î©¥ ÏïÑÎûòÏôÄ Í∞ôÏù¥ Î®ºÏ†Ä ÏÑ§ÏπòÌï¥Ï£ºÏÑ∏Ïöî.

ÏïÑÎûò ÏΩîÎìúÎ•º Ïù¥Ïö©Ìï¥ scikit-imageÏóê ÏÇ¥Í≥†ÏûàÎäî Í≥†ÏñëÏù¥Î•º Ïû†Ïãú Îç∞Î†§ÏòµÎãàÎã§.üê±

````python
from skimage import data
import matplotlib.pyplot as plt

hr_image = data.chelsea() # skimageÏóêÏÑú Ï†úÍ≥µÌïòÎäî ÏòàÏ†ú Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
hr_shape = hr_image.shape[:2]

print(hr_image.shape) # Ïù¥ÎØ∏ÏßÄÏùò ÌÅ¨Í∏∞Î•º Ï∂úÎ†•Ìï©ÎãàÎã§.

plt.figure(figsize=(6,3))
plt.imshow(hr_image)
````

Ïö∞ÏÑ† Î∂àÎü¨Ïò® Í≥†ÏñëÏù¥Í∞Ä ÍΩ§ÎÇò ÏÑ†Î™ÖÌïòÍ∏∞ ÎïåÎ¨∏Ïóê ÏùºÎ∂ÄÎü¨ ÌÅ¨Í∏∞Î•º Ï§ÑÏó¨ Ï†ÄÌï¥ÏÉÅÎèÑÎ°ú ÎÇÆÏ∂∞Î≥¥Í≤†ÏäµÎãàÎã§.   
opencv ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò resize()Î•º Ïù¥Ïö©Ìï¥ Ïù¥ÎØ∏ÏßÄÏùò ÌÅ¨Í∏∞Î•º Ï°∞Ï†àÌï†Ïàò ÏûàÏäµÎãàÎã§.  
ÏïÑÎûò ÏΩîÎìúÏóêÏÑúÎäî Í≥†ÏñëÏù¥Î•º '150 x 100 x 3' ÌÅ¨Í∏∞Î°ú Ï§ÑÏûÖÎãàÎã§.   
dsizeÏùò ÏÑ§Ï†ïÍ∞íÏóê Îî∞Îùº ÌÅ¨Í∏∞Î•º Ï°∞Ï†àÌïòÎäîÎç∞, Ï£ºÏùòÌï† Ï†êÏùÄ Î≥ÄÌôòÌïòÍ≥†Ïûê ÌïòÎäî Ïù¥ÎØ∏ÏßÄÏùò ÌÅ¨Í∏∞Î•º (Í∞ÄÎ°ú ÌîΩÏÖÄ Ïàò, ÏÑ∏Î°ú ÌîΩÏÖÄ Ïàò) Î°ú ÏßÄÏ†ïÌï¥Ï§òÏïº Ìï©ÎãàÎã§.

````python
import cv2
lr_image = cv2.resize(hr_image, dsize=(150,100)) # (Í∞ÄÎ°ú ÌîΩÏÖÄ Ïàò, ÏÑ∏Î°ú ÌîΩÏÖÄ Ïàò)

print(lr_image.shape)

plt.figure(figsize=(3,1))
plt.imshow(lr_image)
````

ÏûëÏïÑÏßÑ Í≥†ÏñëÏù¥ Ïù¥ÎØ∏ÏßÄÎ°ú Interpolation Î∞©Î≤ïÏùÑ Ï†ÅÏö©Ìï¥ Super ResolutionÏùÑ ÏãúÎèÑÌï¥Î≥¥Í≤†ÏäµÎãàÎã§.   resize()ÎÇ¥Ïùò interpolation ÏÑ§Ï†ïÏóê Îî∞Îùº Ï†ÅÏö© Î∞©Î≤ïÏùÑ Ï°∞Ï†àÌï† Ïàò ÏûàÏäµÎãàÎã§.   
ÏïÑÎûò ÏΩîÎìúÎäî Ïù¥Ï†ÑÏóê Í∞ÑÎûµÌïòÍ≤å ÏïåÏïÑÎ≥¥ÏïòÎçò bilinear Î∞è bicubic interpolationÏùÑ Ï†ÅÏö©Ìï¥ Ïù¥ÎØ∏ÏßÄÎ•º  
'300 x 451 x 3' ÌÅ¨Í∏∞Î°ú ÌÅ¨Í≤å Î≥ÄÌôòÌïòÎäî ÏòàÏ†úÏûÖÎãàÎã§.

````python
bilinear_image = cv2.resize(lr_image, 
                            dsize=(451,300), # Ï≤òÏùå Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞Î°ú
                            interpolation=cv2.INTER_LINEAR) # bilinear interpolation Ï†ÅÏö©

bicubic_image = cv2.resize(lr_image,
                          dsize=(451,300), # Ï≤òÏùå Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞Î°ú
                          interpolation=cv2.INTER_CUBIC)

images = [bilinear_image, bicubic_image, hr_image]
titles = ['bilinear', 'bicubic', 'HR']

plt.figure(figsize=(18,18))
for i,(image, title) in enumerate(zip(images, titles)):
    plt.subplot(1,3,i+1)
    plt.imshow(image)
    plt.title(title, fontsize=20)
````

ÌäπÏ†ï Î∂ÄÎ∂ÑÏùÑ ÏûòÎùº Ìï¥ÏÉÅÎèÑ Ï∞®Ïù¥Î•º Î¥êÎ≥¥Ïûê

````python
def crop(image, left_top, x=50, y=100):
    return image[left_top[0]:(left_top[0]+x), left_top[1]:(left_top[1]+y),:]
````

````python
left_tops = [(220,200)] *3 + [(90,120)] *3 + [(30,200)] *3
left_tops
````

````python
plt.figure(figsize=(14,10))
for i, (image, left_top, title) in enumerate(zip(images*3, left_tops, titles*3)):
    plt.subplot(3,3,i+1)
    plt.imshow(crop(image, left_top))
    plt.title(title, fontsize=20)
````

ÏúÑÏôÄ Í∞ôÏùÄ interpolation Î∞©Î≤ïÏùÄ ÎßéÏùÄ Îî•Îü¨Îãù Í∏∞Î∞ò Super Resolution Ïó∞Íµ¨ÏóêÏÑú Í≤∞Í≥ºÎ•º ÎπÑÍµêÌïòÍ∏∞ ÏúÑÌï¥ ÏàòÌñâÎê©ÎãàÎã§.

# 16-8. SRCNNÏùÑ Ïù¥Ïö©Ìï¥ Super Resolution ÎèÑÏ†ÑÌïòÍ∏∞

tensorflow-datasets ÎùºÏù¥Î∏åÎü¨Î¶¨ÏóêÏÑú Ïö∞Î¶¨Í∞Ä ÏÇ¨Ïö©Ìï¥ Î≥º Îç∞Ïù¥ÌÑ∞ÏÖãÏùÄ DIV2K ÏûÖÎãàÎã§  
DIV2K Îç∞Ïù¥ÌÑ∞ÏÖãÏùÄ ÎßéÏùÄ Super Resolution Ïó∞Íµ¨ÏóêÏÑú ÌïôÏäµ Î∞è ÌèâÍ∞ÄÏóê ÏÇ¨Ïö©ÎêòÎäî Îç∞Ïù¥ÌÑ∞ÏÖãÏù¥Î©∞  
800Í∞úÏùò ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞ÏÖã Î∞è 100Í∞úÏùò Í≤ÄÏ¶ùÏö© Îç∞Ïù¥ÌÑ∞ÏÖãÏúºÎ°ú Íµ¨ÏÑ±ÎêòÏñ¥ ÏûàÎã§  
ÏïûÏÑú Îã§Ïö¥Î°úÎìú Ìïú Îç∞Ïù¥ÌÑ∞ÏÖãÏùÄ div2k/bicubic_x4Ïù¥Í≥†  
Ïù¥Îäî DIV2K Îç∞Ïù¥ÌÑ∞ÏÖã Ï§ëÏóêÏÑú Ïã§Ï†ú Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º ÎåÄÏÉÅÏúºÎ°ú bicubic interpolationÏùÑ Ïù¥Ïö©Ìï¥ Í∞ÄÎ°ú Î∞è ÏÑ∏Î°ú ÌîΩÏÖÄ ÏàòÎ•º 1/4Î∞∞Î°ú Ï§ÑÏù∏ Îç∞Ïù¥ÌÑ∞ÏÖã  
Ïù¥Î†áÍ≤å ÎßåÎì§Ïñ¥ÏßÑ Ï†ÄÌï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÏôÄ ÏõêÎûò Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏÑúÎ°ú Ìïú ÏåçÏúºÎ°ú Íµ¨ÏÑ±ÎêòÏñ¥ ÏûàÎã§  
***
ÏïÑÎûòÏóêÏÑú ÏßÑÌñâÌïòÎäî SRCNNÏùò Íµ¨ÌòÑ Î∞è ÌïôÏäµ Í≥ºÏ†ïÏùÄ Îß§Ïö∞ Í∞ÑÎã®Ìïú Í≥ºÏ†ïÎßåÏùÑ Í±∞ÏπòÎ©∞, Ïã§Ï†ú SRCNN ÎÖºÎ¨∏ÏóêÏÑú ÏßÑÌñâÌïú Î∞©ÏãùÍ≥º ÏÑ∏Î∂ÄÏ†ÅÏúºÎ°ú Ï∞®Ïù¥Í∞Ä ÏûàÏäµÎãàÎã§. Ïù¥Ïóê ÎåÄÌïú ÏûêÏÑ∏Ìïú ÎÇ¥Ïö©ÏùÄ SRCNNÏùò ÎÖºÎ¨∏ÏùÑ Ï∞∏Í≥† Î∞îÎûçÎãàÎã§.  
Ï∞∏Í≥† : https://data.vision.ee.ethz.ch/cvl/DIV2K/  


## Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑÌïòÍ∏∞

ÏïÑÎûò ÏΩîÎìúÎ•º Ïù¥Ïö©Ìï¥ DIV2K Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Î∂àÎü¨ÏòµÎãàÎã§.  
ÏïÑÎûò ÏΩîÎìúÏóêÏÑú ÏÉùÏÑ±Ìïú trainÍ≥º validÎäî Í∞ÅÍ∞Å Ï†ÄÌï¥ÏÉÅÎèÑ Î∞è Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º ÏåçÏúºÎ°ú Îã¥Í≥† ÏûàÏäµÎãàÎã§. Í∑∏Ï§ë Ìïú ÏåçÏùò Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌï¥ ÏãúÍ∞ÅÌôîÌï¥Î¥ÖÏãúÎã§.  
Ï†ÄÌï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º bicubic interpolationÌïòÏó¨ Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÏôÄ ÎèôÏùºÌïú ÌÅ¨Í∏∞Î°ú ÎßåÎì§ÏóàÏäµÎãàÎã§.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds

# Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨ÏòµÎãàÎã§.
train, valid = tfds.load(
    "div2k/bicubic_x4", 
    split=["train","validation"],
    as_supervised=True
)

# ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌï¥ Ï≤´Î≤àÏß∏Ïù∏ Ìïú Í∞úÏùò Îç∞Ïù¥ÌÑ∞Îßå ÏÑ†ÌÉùÌï©ÎãàÎã§.
for i, (lr, hr) in enumerate(valid):
    if i == 0: break
    
# Ï†ÄÌï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞Î°ú bicubic interpolation Ìï©ÎãàÎã§.  
hr, lr = np.array(hr), np.array(lr)
bicubic_hr = cv2.resize(
    lr, 
    dsize=(hr.shape[1], hr.shape[0]), # Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞Î°ú ÏÑ§Ï†ï
    interpolation=cv2.INTER_CUBIC # bicubic ÏÑ§Ï†ï
)

# Ï†ÄÌï¥ÏÉÅÎèÑ Î∞è Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.
plt.figure(figsize=(20,20))
plt.subplot(1,2,1); plt.imshow(bicubic_hr) ; plt.title('bicubic_hr', fontsize=25)
plt.subplot(1,2,2); plt.imshow(hr) ; plt.title('hr', fontsize=25)
````

ÏïÑÎûò ÏΩîÎìúÎ•º Ïù¥Ïö©Ìï¥ Í∞Å Ïù¥ÎØ∏ÏßÄÏùò ÌäπÏ†ï Î∂ÄÎ∂ÑÏùÑ ÌôïÎåÄÌï¥Î¥ÖÏãúÎã§.

````python
# Ïù¥ÎØ∏ÏßÄÏùò ÌäπÏ†ï Î∂ÄÎ∂ÑÏùÑ ÏûòÎùºÎÇ¥Îäî Ìï®ÏàòÎ•º Ï†ïÏùòÌï©ÎãàÎã§.
def crop(image, left_top, x=200, y=200):
    return image[left_top[0]:(left_top[0]+x), left_top[1]:(left_top[1]+y), :]

# interpolationÎêú Ïù¥ÎØ∏ÏßÄÏôÄ Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÏùò ÎèôÏùºÌïú Î∂ÄÎ∂ÑÏùÑ Í∞ÅÍ∞Å ÏûòÎùºÎÉÖÎãàÎã§.
left_top = (400, 500)
crop_bicubic_hr = crop(bicubic_hr, left_top)
crop_hr = crop(hr, left_top)

# ÏûòÎùºÎÇ∏ Î∂ÄÎ∂ÑÏùÑ ÏãúÍ∞ÅÌôî Ìï©ÎãàÎã§.
plt.figure(figsize=(15,25))
plt.subplot(1,2,1); plt.imshow(crop_bicubic_hr); plt.title("Bicubic", fontsize=30)
plt.subplot(1,2,2); plt.imshow(crop_hr); plt.title("HR", fontsize=30)
````

SRCNNÏùÑ Ïù¥Ïö©Ìï¥ ÌïôÏäµÌïú Í≤∞Í≥ºÎäî ÏñºÎßàÎÇò Ï¢ãÏùÑÏßÄ ÏïÑÎûòÏóêÏÑú ÌôïÏù∏Ìï¥ Î¥ÖÏãúÎã§  
ÏïûÏÑú Ïñ∏Í∏âÎêòÏóàÎìØÏù¥ SRCNNÏùò ÏûÖÎ†•ÏùÄ Ï†ÄÌï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©ÌïòÎäî Í≤ÉÏù¥ ÏïÑÎãå, ÎßåÎì§Í≥†Ïûê ÌïòÎäî Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞Ïóê ÎßûÍ≤å interpolationÏù¥ Ï†ÅÏö©Îêú Ïù¥ÎØ∏ÏßÄÏûÖÎãàÎã§.  
Ïù¥ Í≥ºÏ†ïÏùÑ ÏïÑÎûò ÏΩîÎìúÏôÄ Í∞ôÏù¥ ÏàòÌñâÌï©ÎãàÎã§.  
ÏÇ¨Ïö©Ìï† DIV2K Îç∞Ïù¥ÌÑ∞ÏÖã ÎÇ¥Ïùò Í∞úÎ≥Ñ Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞Í∞Ä ÌÅ¨ÎØÄÎ°ú, ÏùºÎ∂Ä ÏòÅÏó≠ÏùÑ ÏûÑÏùòÎ°ú ÏûòÎùºÎÇ¥Ïñ¥ ÌïôÏäµÏóê ÌôúÏö©Ìï¥ Î≥¥Í≤†ÏäµÎãàÎã§.

````python
import tensorflow as tf

def preprocessing(lr, hr):
    # Ïù¥ÎØ∏ÏßÄÏùò ÌÅ¨Í∏∞Í∞Ä ÌÅ¨ÎØÄÎ°ú (96,96,3) ÌÅ¨Í∏∞Î°ú ÏûÑÏùò ÏòÅÏó≠ÏùÑ ÏûòÎùºÎÇ¥Ïñ¥ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
    hr = tf.image.random_crop(hr, size=[96, 96, 3])
    hr = tf.cast(hr, tf.float32) / 255.
    
    # ÏûòÎùºÎÇ∏ Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÏùò Í∞ÄÎ°ú, ÏÑ∏Î°ú ÌîΩÏÖÄ ÏàòÎ•º 1/4Î∞∞Î°ú Ï§ÑÏòÄÎã§Í∞Ä
    # interpolationÏùÑ Ïù¥Ïö©Ìï¥ Îã§Ïãú ÏõêÎûò ÌÅ¨Í∏∞Î°ú ÎêòÎèåÎ¶ΩÎãàÎã§.
    # Ïù¥Î†áÍ≤å ÎßåÎì† Ï†ÄÌï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º ÏûÖÎ†•ÏúºÎ°ú ÏÇ¨Ïö©Ìï©ÎãàÎã§.
    lr = tf.image.resize(hr, [96//4, 96//4], "bicubic")
    lr = tf.image.resize(lr, [96, 96], "bicubic")
    return lr, hr

train = train.map(preprocessing).shuffle(buffer_size=10).batch(16)
valid = valid.map(preprocessing).batch(16)
print("‚úÖ")
````

## SRCNN Íµ¨ÌòÑÌïòÍ∏∞

ÏïûÏÑú ÏÇ¥Ìé¥Î≥∏ ÎåÄÎ°ú SRCNNÏùÄ 3Í∞ú convolutional layerÎ•º ÏÇ¨Ïö©Ìïú Íµ¨Ï°∞Î•º Í∞ÄÏßÄÍ≥† ÏûàÏäµÎãàÎã§.  
Îî∞ÎùºÏÑú ÏïÑÎûò ÏΩîÎìúÏôÄ Í∞ôÏù¥ Sequential APIÎ•º ÌôúÏö©Ìï¥ Í∞ÑÎã®ÌïòÍ≤å Íµ¨ÌòÑÌï† Ïàò ÏûàÏäµÎãàÎã§.

````python
from tensorflow.keras import layers, Sequential

# 3Í∞úÏùò convolutional layerÎ•º Í∞ñÎäî Sequential Î™®Îç∏ÏùÑ Íµ¨ÏÑ±Ìï©ÎãàÎã§.
srcnn = Sequential()
# 9x9 ÌÅ¨Í∏∞Ïùò ÌïÑÌÑ∞Î•º 128Í∞ú ÏÇ¨Ïö©Ìï©ÎãàÎã§.
srcnn.add(layers.Conv2D(128, 9, padding="same", input_shape=(None, None, 3)))
srcnn.add(layers.ReLU())
# 5x5 ÌÅ¨Í∏∞Ïùò ÌïÑÌÑ∞Î•º 64Í∞ú ÏÇ¨Ïö©Ìï©ÎãàÎã§.
srcnn.add(layers.Conv2D(64, 5, padding="same"))
srcnn.add(layers.ReLU())
# 5x5 ÌÅ¨Í∏∞Ïùò ÌïÑÌÑ∞Î•º 3Í∞ú ÏÇ¨Ïö©Ìï©ÎãàÎã§.
srcnn.add(layers.Conv2D(3, 5, padding="same"))

srcnn.summary()
````

## SRCNN ÌïôÏäµÌïòÍ∏∞

ÏµúÏ†ÅÌôî Î∞©Î≤ïÍ≥º ÏÜêÏã§ Ìï®ÏàòÎ•º ÏÑ§Ï†ïÌïòÍ≥†, ÏïûÏÑú Ï†ÑÏ≤òÎ¶¨Ìïú Îç∞Ïù¥ÌÑ∞Î•º Ïù¥Ïö©Ìï¥ ÏïÑÎûòÏôÄ Í∞ôÏù¥ Í∞ÑÎã®ÌïòÍ≤å ÌïôÏäµÌï† Ïàò ÏûàÏäµÎãàÎã§.  
Ïö∞ÏÑ† 1 epochÎßå ÌïôÏäµ ÏãúÏºú Î¥ÖÏãúÎã§.

````python
srcnn.compile(
    optimizer="adam", 
    loss="mse"
)

srcnn.fit(train, validation_data=valid, epochs=1)
````

## SRCNN ÌÖåÏä§Ìä∏ÌïòÍ∏∞

SRCNNÏùò ÌïôÏäµÏóêÎäî ÍΩ§ÎÇò Ïò§ÎûúÏãúÍ∞ÑÏù¥ ÏÜåÏöîÎêòÏñ¥ Ïó¨Í∏∞ÏÑú SRCNNÏùÑ Ï≤òÏùåÎ∂ÄÌÑ∞ ÎÅùÍπåÏßÄ ÌïôÏäµÏãúÌÇ§ÏßÄ ÏïäÍ≤†ÏäµÎãàÎã§. ÎåÄÏã† SRCNNÏùÑ ÏÇ¨Ïö©Ìï¥Î≥¥Í∏∞ ÏúÑÌï¥ Ïù¥ÎØ∏ ÌïôÏäµÏù¥ ÏôÑÎ£åÎêú SRCNN Î™®Îç∏ÏùÑ Ï§ÄÎπÑÌñàÏäµÎãàÎã§üòÖ ÏïÑÎûò srcnn.h5 ÌååÏùºÏùÑ Îã§Ïö¥Î°úÎìúÎ∞õÏùÄ ÌõÑ ÏΩîÎìúÎ•º Ïã§ÌñâÌïòÏó¨ ÌïôÏäµ ÏôÑÎ£åÎêú Î™®Îç∏ÏùÑ Î∂àÎü¨Ïò¨ Ïàò ÏûàÏäµÎãàÎã§.

```
$ mv srcnn.h5 ~/aiffel/super_resolution
```  

````python
import tensorflow as tf
import os

model_file = os.getenv('HOME')+'/aiffel/super_resolution/srcnn.h5'
srcnn = tf.keras.models.load_model(model_file)
````

ÏïÑÎûò ÏΩîÎìúÏôÄ Í∞ôÏù¥ Ï†ÄÌï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º ÏûÖÎ†•Î∞õÏïÑ SRCNNÏùÑ ÏÇ¨Ïö©ÌïòÎäî Ìï®ÏàòÎ•º Í∞ÑÎã®ÌïòÍ≤å Ï†ïÏùòÌïòÍ≥†,  
Ïù¥ Ìï®ÏàòÎ•º Ïù¥Ïö©Ìï¥ SRCNNÏùò Í≤∞Í≥ºÏù∏ Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÎ•º ÏñªÏñ¥Î¥ÖÏãúÎã§.

````python
def apply_srcnn(image):
    sr = srcnn.predict(image[np.newaxis, ...]/255.)
    sr[sr > 1] = 1
    sr[sr < 0] = 0
    sr *= 255.
    return np.array(sr[0].astype(np.uint8))

srcnn_hr = apply_srcnn(bicubic_hr)
````

Ïù¥Ï†Ñ bicubic interpolationÏùÑ Ï†ÅÏö©ÌïòÍ≥† Ïù¥ÎØ∏ÏßÄ Ï†ÑÏ≤¥Î•º ÏãúÍ∞ÅÌôîÌñàÏùÑ Îïå ÏÑ∏Î∂ÄÏ†ÅÏù∏ ÏÑ†Î™ÖÌï®Ïù¥ ÎààÏóê ÎùÑÏßÄ ÏïäÏïòÍ∏∞ ÎïåÎ¨∏Ïóê, ÏùºÎ∂Ä ÏòÅÏó≠ÏùÑ ÏûòÎùºÎÇ¥Ïñ¥ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÎπÑÍµêÌï¥Î≥¥Í≤†ÏäµÎãàÎã§.  
ÏïÑÎûò ÏΩîÎìúÎ•º Ïù¥Ïö©Ìï¥ 3Í∞ú Ïù¥ÎØ∏ÏßÄ(bicubic interpolationÏùò Í≤∞Í≥º, SRCNNÏùò Í≤∞Í≥º, ÏõêÎûò Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄ)Î•º ÎÇòÎûÄÌûà ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.

````python
# ÏûêÏÑ∏Ìûà ÏãúÍ∞ÅÌôî ÌïòÍ∏∞ ÏúÑÌï¥ 3Í∞ú ÏòÅÏó≠ÏùÑ ÏûòÎùºÎÉÖÎãàÎã§.
# ÏïÑÎûòÎäî ÏûòÎùºÎÇ∏ Î∂ÄÎ∂ÑÏùò Ï¢åÏÉÅÎã® Ï¢åÌëú 3Í∞ú ÏûÖÎãàÎã§.
left_tops = [(400,500), (300,1200), (0,1000)]

images = []
for left_top in left_tops:
    img1 = crop(bicubic_hr, left_top, 200, 200)
    img2 = crop(srcnn_hr , left_top, 200, 200)
    img3 = crop(hr, left_top, 200, 200)
    images.extend([img1, img2, img3])

labels = ["Bicubic", "SRCNN", "HR"] * 3

plt.figure(figsize=(18,18))
for i in range(9):
    plt.subplot(3,3,i+1) 
    plt.imshow(images[i])
    plt.title(labels[i], fontsize=30)
````

ÏãúÍ∞ÅÌôî Í≤∞Í≥º bicubic interpolation Í≤∞Í≥ºÎ≥¥Îã§ Ï°∞Í∏à Îçî ÏÑ†Î™ÖÌï¥Ï°åÏßÄÎßå ÏõêÎûò Í≥†Ìï¥ÏÉÅÎèÑ Ïù¥ÎØ∏ÏßÄÏóê ÎπÑÌï¥ ÎßåÏ°±Ìï†ÎßåÌïú ÏÑ±Îä•ÏùÄ ÏïÑÎãå Í≤É Í∞ôÏäµÎãàÎã§. DIV2K Îç∞Ïù¥ÌÑ∞ÏÖãÏù¥ ÎπÑÍµêÏ†Å ÏÑ∏Î∞ÄÌïú Íµ¨Ï°∞Ïùò Ïù¥ÎØ∏ÏßÄÍ∞Ä ÎßéÏïÑ SRCNNÍ≥º Í∞ôÏù¥ Í∞ÑÎã®Ìïú Íµ¨Ï°∞Î°úÎäî Îçî Ïù¥ÏÉÅ ÌïôÏäµÎêòÏßÄ ÏïäÎäî Í≤ÉÏúºÎ°ú Î≥¥ÏûÖÎãàÎã§.  

Ïã§Ï†úÎ°ú SRCNN ÎÖºÎ¨∏ÏóêÏÑúÎäî ÏïÑÎûòÏôÄ Í∞ôÏù¥ ÎπÑÍµêÏ†Å Í∞ÑÎã®Ìïú Íµ¨Ï°∞Ïùò Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌï¥ÏÑúÎäî ÍΩ§ÎÇò ÎßåÏ°±Ìï† ÎßåÌïú ÏÑ±Îä•ÏùÑ Î≥¥Ïó¨Ï§çÎãàÎã§.
